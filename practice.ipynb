# Functions in Python

A function is a reusable block of code that performs a specific task.
You define it once and call it whenever you need it.

## Simple Explanation
A function in Python is a reusable piece of code defined with the `def` keyword. After defining it, you can call it whenever you need to run that task.

````python
# defines the function.
def greet():
    print("Hello!")
# call the function:
greet()
````

````python
def add(a,b):
    return a + b
# call the function:
result = add(3 , 5)
print (result)


````

## Step-by-Step: How to Create a Function

### 1. Decide the Task
Ask yourself: What should this function do?

Examples:
- Say hello
- Add two numbers
- Check if a number is even
- Calculate area

A function should usually do one clear task.

### 2. Choose a Good Name
Pick a name that explains the job.

Good names: `say_hello`, `add_numbers`, `check_even`, `calculate_area`
Less clear names: `doit`, `x1`, `test`

### 3. Decide the Inputs (Parameters)
Ask: What information does the function need?

Examples:
- Add numbers: `a`, `b`
- Greet someone: `name`
- Rectangle area: `length`, `width`

### 4. Write `def`
Every function starts with `def`.

```python
def say_hello():
    print("Hello")
```

### 5. Indent the Function Body
Everything inside the function must be indented.

### 6. Call the Function
Defining a function is not enough. You must call it.

```python
def say_hello():
    print("Hello")

say_hello()
```

Output: `Hello`

````python
def greet(name):
    print("Hello", name)

greet("Robabeh")
````

## `print()` vs `return`

This is one of the most important function concepts.

- `print()` displays output on the screen.
- `return` sends a value back so you can store or reuse it later.

### Example with `print()`
In the next code cell, the function prints the result but does not return it.

````python
def add_numbers(a, b):
    print(a + b)

result = add_numbers(2, 3)
print(result)
````

### Why do you see `None`?

Because the function printed the answer, but did not `return` a value.
When no value is returned, Python returns `None` by default.

````python
def add_numbers(a, b):
    return a + b

result = add_numbers(2, 3)
print(result)
````

## Example 1: Check if a Number Is Even

### Task
Check whether a number is even.

### Input
One number.

### Output
`True` or `False`.

### Code
See the next code cell.

````python
def is_even(number):
    return number % 2 == 0  # == (Comparison Operator) 
# Call it to check if a number is even or not:
print(is_even(4))  # Output: True
print(is_even(7))  # Output: False  
 # % means remainder
#even numbers have remainder 0 when divided by 2
````

````python
# Example 2: Calculate rectangle area

def calculate_area( length, width):
    area = length * width
    return area
print(calculate_area(5, 4))
````
